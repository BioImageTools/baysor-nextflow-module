```python
def fov_region(cells: list, xmin, xmax, ymin, ymax):
    #x_tuple = (xmin, )
    fov_cells = []
    for cell in cells:
        cell_xmin, cell_xmax = np.min(cell[:, 0]), np.max(cell[:, 0])
        cell_ymin, cell_ymax = np.min(cell[:, 1]), np.max(cell[:, 1])
        if (cell_xmin > xmin) & (cell_xmax < xmax):
            fov_cells.append(cell)
    return fov_cells
```
